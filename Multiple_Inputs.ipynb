```python
from typing import Dict, TypedDict, List
from langgraph.graph import StateGraph # framework that helps you design and manage the flow of tasks in your application using graphs 
```

```python
class AgentState(TypedDict):
    values : List[int]
    name : str
    result : str
```

```python
def process_values(state: AgentState) -> AgentState:
    """This function handles multiple different inputs."""
    state['result'] = f"Hi there {state['name']}, the sum of your values is {sum(state['values'])}."
    
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("processer",process_values)
graph.set_entry_point("processer")
graph.set_finish_point("processer")
app = graph.compile()
```

```python
result = app.invoke({
    "values": [1,2,3,4,5],
    "name": "Aum"
})
result['result']
```
