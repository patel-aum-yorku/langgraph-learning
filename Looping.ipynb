```python
from typing import Dict, TypedDict, List
from langgraph.graph import StateGraph, START, END
import random

class AgentState(TypedDict):
    name: str
    number: list[int]
    coutner: int
```

```python
def greeting_node(state:AgentState) -> AgentState:
    """This node greets the user."""
    state["name"] = f"Hi there! My name is {state['name']}."
    state["coutner"] = 0
    return state

def random_node(state:AgentState) -> AgentState:
    """This node generates a random number and appends it to the list."""
    rand_num = random.randint(1, 10)
    state["number"].append(rand_num)
    state["coutner"] += 1
    return state

def should_continue(state:AgentState) -> AgentState:
    """This node checks if we should continue looping."""
    if state["coutner"] < 5:
        print("Entering the loop", state["coutner"])
        return "loop"
    else:
        return "exit"
```

```python
graph = StateGraph(AgentState)
graph.add_node("greeting", greeting_node)
graph.add_node("random", random_node)
graph.add_edge
```
